# 数値解析 前半課題

1. 掃出し法を用いて，次の連立方程式を解きなさい．

$$ -3{y} + 2{z} + {w} = 7 $$
$$ 3{x} + 2{y} - 3{z} = -1 $$
$$ {x} + 2{y} - 3{z} + 2{w} = 3 $$
$$ -3{x} + 4{y} + {z} + 2{w} = 9$$

<br>

2. ガウス・ザイデル法を用いて，次の連立方程式を解きなさい．

$$ 4{x_1} - {x_2} - {x_5} = 3 $$
$$ -{x_1} + 4{x_2} - {x_3} = 3 $$
$$ -{x_2} + 4{x_3} - {x_4} = 7 $$
$$ -{x_3} + 4{x_4} - {x_5} = 9 $$
$$ -{x_1} - {x_4} + 4{x_5} = 8 $$

<br>

3. はさみうち法を用いて，次の方程式の1根（2.7に近いλの値）を求めなさい．

$$ 1.25\sin(0.4{λ})\sin(0.6{λ})-{λ}\sin{λ} = 0 $$

<br>

4. ニュートン法を用いて，次の方程式の1根を求めなさい．

$$ {x^2} - 2 = 0 $$

<br>
